## Methods
### Data Collection

This dataset is maintained publicly at [Kaggle](https://www.kaggle.com/datasets/catherinerasgaitis/mxmh-survey-results), a free online resource of general datasets [1].

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load data
MusicData = pd.read_csv('https://raw.githubusercontent.com/eromoselekelvinsmart/Assignment-1/main/mxmh_survey_results.csv')
print(MusicData.shape)
MusicData
```

### Data Cleaning 
The dataset had a lot of unwanted colums that made the data biased and some hard to understand rows and columns so i renamed the rows, columns and reduced to the colums to a needed amount of colunms 

```python
print(f"Null values of the Dataset :")
MusicData.isna().sum().to_frame()
```

They are some null values in the data set so i removed them

```python
MusicData = MusicData.dropna()
MusicData
```

```python
MusicData = MusicData.drop(['Timestamp', 'Primary streaming service', 'Foreign languages', 'Permissions', 'BPM'], axis=1)
MusicData.tail()
```

## Analyses
---
I cleaned my data and removed the unwanted columns that weren't helping my analysis. The goal of this analysis is to find out the relationship of music and mental health issues and other outliers that might support or deter the notion. i will find the relationship between the amount of music that is heard and mental health issues then i will determine if age affects the effiency of the "Music Healing" and finally, i will determine if exploring other music helps mental health 

```python
print(f"Summary Of The data set :")
MusicData.describe()
```

```python
h = MusicData.Age.hist(bins=8, figsize=(7,5))
```

This shows that most of the people in this survey are 10 - 40

```python
grouped_data = MusicData.groupby(['Music effects','Hours per day','Anxiety']).size().unstack().fillna(value= 0)
grouped_data #to understand the degree of errors

sns.set(rc={"figure.figsize":(15, 5)})
sns.barplot(x = "Hours per day", y = "Anxiety", hue = "Music effects", data = MusicData)
```

We can conclude that the more music they litsen to seems to improve their anxiety and less or no music ha no effects or worsened the case


To be more precise, i will test the relationships with other mental case issues.

```python
grouped_data = MusicData.groupby(['Music effects','Hours per day','Depression']).size().unstack().fillna(value= 0)
grouped_data #to understand the degree of errors 

sns.set(rc={"figure.figsize":(15, 5)})
sns.barplot(x = "Hours per day", y = "Depression", hue = "Music effects", data = MusicData)
```

```python
grouped_data = MusicData.groupby(['Music effects','Hours per day','Insomnia']).size().unstack().fillna(value= 0)
grouped_data #to understand the degree of errors

sns.set(rc={"figure.figsize":(15, 5)})
sns.barplot(x = "Hours per day", y = "Insomnia", hue = "Music effects", data = MusicData)
```

```python
grouped_data = MusicData.groupby(['Music effects','Hours per day','OCD']).size().unstack().fillna(value= 0)
grouped_data #to understand the degree of errors

sns.set(rc={"figure.figsize":(15, 5)})
sns.barplot(x = "Hours per day", y = "OCD", hue = "Music effects", data = MusicData)
```

---
As shown, the case is the same for other mental health issues, so we can conclude that music improves your mental health state.
I will now determine the song genre that improves the mental health the most.

```python
grouped_data = MusicData.groupby(['Hours per day', 'Fav genre', 'Music effects']).size().unstack().fillna(value= 0)
grouped_data #to understand the degree of errors

sns.set(rc={"figure.figsize":(19, 6)})
sns.barplot(x = "Fav genre", y = "Hours per day", hue = "Music effects", data = MusicData)
```

The values are far to diffrent to have a relationship with each other, therefore we can conclude that the genre of music has little effect on the mental health.

To identify the reason music helps with mental health i will determine if age affects the effiency of the "Music Healing".

```python
sns.regplot(x=MusicData['Age'], y=MusicData['Hours per day'])
MusicData['Age'].corr(MusicData['Hours per day'])
```

As seen above there is no correlation between age and the amount of music listened to. We have proof that the hours of music listened to per day affects the "Music healing" so as such we can imply that age does not relate but is that statement true.
To determine if the statemnet is true we have to make another graph and analyze it.

```python
sns.set(rc={"figure.figsize":(25, 8)})
sns.countplot(data=MusicData, x="Age", hue="Music effects").set(title='The degree of musics effects on age')
```

This graph contadicts the statement made previously so we can say that age does have an effect on music healing but the older one gets the worse the efficency.

```python
sns.countplot(data=MusicData, x="Exploratory", hue="Music effects")
```

We can see that exploring new music other than your personal taste has a huge effect on the effect on "Music Healing"
